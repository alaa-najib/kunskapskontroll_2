### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure: Det är hela ytan där diagram ritas, som en tom duk.
# Axes: Det är själva diagrammet inom figuren där datan visas.
# Axis: Det är x- och y-axlarna som visar skalor och hjälper att förstå datan.
# Artist: Allt som syns i diagrammet, som linjer, text eller punkter, kallas en artist.
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5, label="Random Data")
ax1.plot(x, y, color="red", label="y = x^2")
ax1.set_title("Scatter Plot")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.legend

# Bar plot
ax2.bar(names, values, color="orange")
ax2.set_title("Bar Plot")
ax2.set_xlabel("Fruits")
ax2.set_ylabel("Values")

# Show plot
plt.tight_layout()
plt.show()
```
